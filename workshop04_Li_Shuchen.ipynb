Exercise - Google Play Store
1. Number of app in each category
2. Average ratings for each category
3. For the category with the highest rating, 
4. plot the ratings in the following ranges - 5 to 4.5, 4.5 to 3.5, 3.5 to 2.5, 2.5 and below (edited) 

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('googleplaystore.csv') #"https://chukmunnlee.github.io/dataset/googleplaystore.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
mask = df.Category != "1.9"
df2 = df[mask].copy()
df2.Category.value_counts().sort_index().plot(kind="bar", figsize= (12,8), label ="category count", grid=True)
plt.legend()
```

```python
df.dropna(inplace=True)
```

```python
df.groupby("Category")["Rating"].mean().plot(kind="bar",figsize= (12,8), grid = True, legend = "Avg Rating")
```

```python
#Plot out diff rangers.. 
# 5 - 4.5 
# 4.5 - 3.5
# 3.5 - 2.5
# 2.5
```

```python
mask2 = df.Category == "FAMILY"
df3 = df[mask2]
x = ["2.5-0", "3.5-2.5", "4.5-3.5", "5-4.5"]
bar = [
    len(df3[df3.Rating.between(0, 2.5)]), 
    len(df3[df3.Rating.between(2.5, 3.5)]),
    len(df3[df3.Rating.between(3.5, 4.5)]),
    len(df3[df3.Rating.between(4.5, 5)])
]

width = 1/1.5
plt.figure (figsize = (12, 8))
plt.bar(x,bar,width,color="navy", label = "Rating group counts")   
plt.legend ()
plt.grid()
```
